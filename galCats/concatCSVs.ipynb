```python
import pandas as pd
import os
import glob
```

```python
allDirs = glob.glob("/global/cfs/cdirs/lsst/groups/MCP/standardSirens/mockGalCats/prod_*/")
```

```python
for input_folder in allDirs:
    print(f"Processing .csv's in {input_folder}...")
    
    output_file = os.path.join(input_folder,'combined.csv')
    chunk_size = 5000000  # Process 5M rows at a time to keep RAM low
    
    files = [f for f in os.listdir(input_folder) if f.endswith('.csv')]
    first_file = True

    print(f"All files in {input_folder}:")
    for f in files:
        print(f)

    for filename in files:
        file_path = os.path.join(input_folder, filename)
        print(f"Processing {filename}...")
        
        # Read the file in chunks
        for chunk in pd.read_csv(file_path, chunksize=chunk_size):
            if first_file:
                # 1. Establish the Master Header from the very first chunk
                master_columns = chunk.columns.tolist()
                chunk.to_csv(output_file, index=False, mode='w')
                first_file = False
            else:
                # 2. Reorder this chunk to match the Master Header
                # Any missing columns will be filled with NaN
                chunk = chunk.reindex(columns=master_columns)
                chunk.to_csv(output_file, index=False, mode='a', header=False)
    
    print(f"Finished! Combined data saved to {output_file}")
```
