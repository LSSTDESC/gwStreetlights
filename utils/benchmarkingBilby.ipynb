```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n_nodes = np.array([1,2,4],dtype=int)
coresPerNode = np.array([256,256,256])
wallClockTime = np.array([np.nan, np.nan,21923]) # in seconds
cpuEfficiency = np.array([np.nan, np.nan,20.74]) # Percent
memoryUtilized = np.array([np.nan, np.nan,56.94]) # in GB
memoryEfficiency = np.array([np.nan, np.nan,2.99]) # Percent, out of 1.99TB
jobIDs = np.array([37014170,37014174,36980979])
```

```python
benchmarkDF = pd.DataFrame([n_nodes,
                            coresPerNode,
                            wallClockTime,
                            cpuEfficiency,
                            memoryUtilized,
                            memoryEfficiency],
                            columns=jobIDs,index=["n_nodes",
                                                  "cores_per_node",
                                                  "Wall Clock Time [s]",
                                                  "CPU Efficiency [%]",
                                                  "Memory Utilized [GB]",
                                                  "Memory Efficiency [%]"]).T
```

```python
fig,axs = plt.subplots(2,2,figsize=[10,10],sharex=True)
fig.suptitle("Request statistics for 256 cores per node")
for column,ax in zip(benchmarkDF.columns.values[2:],axs.flatten()):
    ax.scatter(benchmarkDF["n_nodes"],benchmarkDF[column])
    ax.set_ylabel(column)
    ax.grid(ls='--')
    
for ax in axs[1,:]:
    ax.set_xlabel("Nodes requested")
fig.tight_layout()
```
